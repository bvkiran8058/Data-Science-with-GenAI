```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_excel('../data/titanic3.xls')
dataset.head()
```

## distribution plot

```python
sns.displot(data=dataset, x='age', bins=20, edgecolor='black', hue='survived')
plt.grid(True)
plt.show()
```

## joint plot

```python
sns.jointplot(data=dataset, x='age', y='fare', hue='survived')
plt.grid(True)
plt.show()
```
