```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_excel('../data/titanic3.xls')
dataset.head()
```

## distribution plot

```python
sns.displot(data=dataset, x='age', bins=20, edgecolor='black', hue='survived')
plt.grid(True)
plt.show()
```

## joint plot

```python
sns.jointplot(data=dataset, x='age', y='fare', hue='survived')
plt.grid(True)
plt.show()
```

## pair plot

```python
sns.pairplot(dataset, hue='survived')
plt.figure(figsize=(5,5))
plt.show()
```

## barplot

```python
sns.barplot(x='sex', y='fare', data=dataset, estimator='mean', hue='survived')
plt.title('Average Fare by Sex')
plt.show()
```

## countplot

```python
sns.countplot(x='pclass', hue='sex', data=dataset)
plt.title('Passenger Count by Class and Sex')
plt.show()
```

## boxplot

```python
sns.boxplot(x='pclass', y='age', hue='sex', data=dataset)
plt.title('Age Distribution by Class and Sex')
plt.show()
```

## heatmap

```python
corr = dataset[['age', 'fare', 'pclass']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

## pairgrid

```python
g = sns.PairGrid(dataset[['age', 'fare', 'pclass']])
g.map(sns.scatterplot)
plt.show()
```
