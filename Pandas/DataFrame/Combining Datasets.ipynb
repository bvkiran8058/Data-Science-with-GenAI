## Concatenate

```python
data1 = {
    'a':[1,2,3,4],
    'b':[5,6,7,8],
    'c':[9,10,11,12]
}
data2={
    'a':[13,14,15,16],
    'b':[17,18,19,20],
    'c':[21,22,23,24],
    'd':[25,26,27,28]
}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
df2
```

```python
pd.concat((df1, df2)) # by default axis =0 so combined with rows
```

```python
pd.concat((df1, df2), axis=1)
```

```python
pd.concat((df1, df2), keys= ['data1', 'data2'])
```

```python
pd.concat((df1, df2), keys= ['data1', 'data2'], axis=1)
```

## merge

```python
use_cols = ['Name', 'Salary' ]
```

```python
company_21 = pd.read_csv('../../data/company21.csv', sep=';', usecols=use_cols)
company_22 = pd.read_csv('../../data/company22.csv', sep=';', usecols=use_cols)
```

```python
company_21
```

```python
company_22
```

```python
pd.merge(
    left=company_21, 
    right=company_22,
    how='outer',
    on='Name',
    indicator=True,
) # by change how we can merge in different ways
# how : {'left', 'right', 'outer', 'inner', 'cross'}, default 'inner'
#     Type of merge to be performed.

# left: use only keys from left frame, similar to a SQL left outer join; preserve key order.
# right: use only keys from right frame, similar to a SQL right outer join; preserve key order.
# outer: use union of keys from both frames, similar to a SQL full outer join; sort keys lexicographically.
# inner: use intersection of keys from both frames, similar to a SQL inner join; preserve the order of the left keys.
# cross: creates the cartesian product from both frames, preserves the order of the left keys.
```
