## apply

```python
pokemon['name_length'] = pokemon['Name'].apply(len)
pokemon.head()
```

```python
# own function
def strength(value : int) -> str:
    if value > 150:
        return 'Strong'
    elif value > 100:
        return 'Average'
    else:
        return 'Weak'
```

```python
pokemon['Strength'] = pokemon['Attack'].apply(strength)
pokemon.head()
```

```python
pokemon['Strength'].value_counts()
```
