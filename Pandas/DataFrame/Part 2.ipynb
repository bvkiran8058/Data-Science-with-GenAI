## broadcasting
* this concept is taken from numpy
* if we are performing any operation with different shapes it will automatically replace that value with NaN

```python
a = pd.Series([1,2,3,4])
b = pd.Series([1,2,3,4,5])
a+b
```

```python
a = pd.Series([1,2,3,4])
b = pd.Series([1,2,3,4])
a+b
```

```python
df1 = pd.DataFrame(data={
    'a':a,
    'b':b
})
df2 = pd.DataFrame(data={
    'a':a,
    'b':b
})
df1, df2, df1+df2
```

```python
df1 = pd.DataFrame(data={
    'a':a,
    'b':b
})
df2 = pd.DataFrame(data={
    'a':[1,2,3,4,5],
    'b':[1,2,3,4,5]
})
df1, df2, df1+df2
```

## filtering with conditions

### single condtion

```python
titanic['sex']=='male'
```

```python
male_mask = titanic['sex']=='male'
print((male_mask))
```

```python
titanic[titanic['sex']=='male'].head()
```

```python
titanic[male_mask].head()
```

```python
children = titanic['age'] <= 18
children
```

```python
titanic[children].head()
```

### multiple conditions

```python
titanic[male_mask & children]
```

```python
# filter female and survived and aged above 60
female_masked = titanic['sex'] == 'female'
survived = titanic['survived'] == 1
aged = titanic['age'] >= 60
titanic[female_masked & survived & aged]
```

## isin, isnull, notnull

```python
first_class = titanic['pclass'] == 1
second_class = titanic['pclass'] == 2
titanic[first_class | second_class].shape
```

```python
titanic['pclass'].isin([1,2])
```

```python
titanic[titanic['pclass'].isin([1,2])].shape
```

```python
titanic['age'].isnull()
```

```python
titanic[~ titanic['age'].isnull()] # known age ur use notnull
```
