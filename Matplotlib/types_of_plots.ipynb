```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Scatter plots

```python
x = np.random.random(100) * 100
y1 = np.random.random(100) * 100
colors = np.random.random(100) * 100
sizes = np.random.random(100) * 100
plt.scatter(x, y1, c= colors, alpha=0.5, cmap='hot')
plt.title('random scatter plot')
plt.xlabel('X')
plt.ylabel('Y values')
plt.colorbar()
plt.show()
```

```python
x = np.random.random(100) * 100
y1 = np.random.random(100) * 100
y2 = np.random.random(100) * 100
y3 = np.random.random(100) * 100

plt.scatter(x, y1, color='r', alpha=0.6, label='y1')
plt.scatter(x, y2, color='g', alpha=0.6, label='y2')
plt.scatter(x, y3, color='b', alpha=0.6, label='y3')

plt.title('Multiple 2D Scatter Plots')
plt.xlabel('X')
plt.ylabel('Y values')
plt.legend(loc='best')
plt.show()
```

## Line plots

```python
x = list(range(0, 100, 2))
y1 = np.random.randint(0,100,50)
y2 = np.random.randint(100,200,50)
# y1 = np.linspace(100, 0, 50)
# y2 = np.linspace(0, 100, 50)
plt.figure(figsize=(10,5))
plt.plot(x, y1, c='b', label='y1')
plt.plot(x, y2, c='r', label='y2')
plt.legend(loc='best')
plt.show()
```

## barplots

```python
prog_langs = ['c', 'c++', 'c#', 'java', 'python']
students = [89, 61, 70, 20, 100]
plt.bar(prog_langs, students, width=0.7, color=['skyblue', 'b', 'orange', 'purple', 'g'])
plt.xlabel("prog_langs")
plt.ylabel('no of students')
plt.title("learnings in a class")
plt.legend()
plt.show()
```

## histogram

```python
sample = np.random.randint(0,50,50)
print(np.bincount(sample).argmax())
plt.hist(sample, bins = 20, histtype='barstacked', rwidth=0.5)
```

## piecharts

```python
prog_langs = ['c', 'c++', 'c#', 'java', 'python']
students = [89, 61, 70, 20, 100]
plt.pie(x=students, explode=[0, 0, 0, 0, 0.2], labels=prog_langs, colors=['skyblue', 'b', 'orange', 'purple', 'g'], autopct='%.3f', pctdistance=1.3)
plt.show()
```
